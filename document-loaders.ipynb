```python
from langchain_community.document_loaders.csv_loader import CSVLoader

file_path = (
    "/kaggle/input/test-ds/annual-enterprise-survey-2023-financial-year-provisional.csv"
)

loader = CSVLoader(file_path=file_path)
data = loader.load()

for record in data[:1]:
    print(record)
```

**Load with options**

```python
loader = CSVLoader(
    file_path=file_path,
    csv_args={
        "delimiter": ",",
        "quotechar": '"'
    },
)

data = loader.load()
print(data[2])
for record in data[:2]:
    print(record)
```
